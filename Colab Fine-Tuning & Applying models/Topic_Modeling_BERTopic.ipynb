```python
##Perform BERTopic on the "lemmas" column of the dataset
from bertopic import BERTopic
from sklearn.datasets import fetch_20newsgroups
import pandas as pd

df1 = pd.read_csv('/content/drive/MyDrive/data_stopwords_refugee1.tsv', sep='\t', header = 0)

# Ensure that docs variable is list of strings
docs = df1["lemmas"].dropna().tolist()

# Ensure there are no empty strings
docs = [doc for doc in docs if doc != ""]

# Initialize the model
topic_model = BERTopic(language="multilingual")

# Fit the model
topics, probs = topic_model.fit_transform(docs)

# Retrieve information about the topics from the topic model
topics_infos = topic_model.get_topic_info()
```

```python
#Calculating the Coherence Score
from gensim.corpora.dictionary import Dictionary
from gensim.models.coherencemodel import CoherenceModel
import ast

# convert string representation of list to list
docs = [ast.literal_eval(doc) for doc in docs]

# Create a dictionary representation of the documents.
dictionary = Dictionary(docs)

# Filter out words that occur less than 20 documents, or more than 50% of the documents.
dictionary.filter_extremes(no_below=20, no_above=0.5)

# Create bag-of-words representation of the documents.
corpus = [dictionary.doc2bow(doc) for doc in docs]

# Get topics from BERTopic
topic_words = [topic_model.get_topic(topic) for topic in set(topics) if topic != -1]

# Use only the words from the topics
topic_words = [[word for word, _ in words] for words in topic_words]

# Compute Coherence Score using c_v
coherence_model = CoherenceModel(topics=topic_words, texts=docs, dictionary=dictionary, coherence='c_v')
coherence_score = coherence_model.get_coherence()

print('\nCoherence Score: ', coherence_score)
```

```python
#Calculating the silhouette score
from sklearn.metrics import silhouette_score

# Extract embeddings from BERTopic
embeddings = topic_model._extract_embeddings(docs)

# Predict topics
predictions = topic_model.transform(docs)

# The transform function returns two values: topics and probabilities
# We only need topics for silhouette score
topics, _ = predictions

# Calculate silhouette score
sil_score = silhouette_score(embeddings, topics)

print("Silhouette score: ", sil_score)
```
