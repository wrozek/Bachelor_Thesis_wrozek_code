```python
# Import the necessary libraries and modules
from transformers import AutoTokenizer, AutoModelForTokenClassification, Trainer, TrainingArguments
from datasets import load_dataset
```

```python
#Load dataset
dataset = load_dataset('json', data_files='allsentiment4.jsonl')
```

```python
#Print dataset for test purposes
print(dataset)
```

```python
#Print the first instance of the dataset to see how it looks like
dataset["train"][0]
```

```python
#Remove unnecessary columns
dataset = dataset.remove_columns("id")
dataset = dataset.remove_columns("Comments")
```

```python
# Mapping labels to single values
def map_labels(example):
    example['label'] = example['label'][0]
    return example
```

```python
# Mapping labels for the 'train' dataset
dataset['train'] = dataset['train'].map(map_labels)
```

```python
# Import the necessary libraries and modules
from datasets import load_dataset, Features, Value, ClassLabel, load_metric
from transformers import AutoTokenizer
from transformers import AutoModelForTokenClassification, AutoModelForSequenceClassification, TrainingArguments, Trainer
from transformers import DataCollatorForTokenClassification
import numpy as np
#Check if CUDA is available
import torch
print(torch.cuda.is_available())
```

```python
# Modifying features for the 'train' dataset
features = dataset["train"].features.copy()
features["label"] = ClassLabel(names=["negative", "neutral", "positive"])
```

```python
# Adjusting labels to numerical values
label_dict = {
    "negative" : 1,
    "neutral" : 2,
    "positive" : 0
}

def adjust_labels(batch):
    batch["label"] = [label_dict[sentiment] for sentiment in batch["label"]]
    return batch
```

```python
# Mapping sentiment labels to numerical values in the training dataset
dataset["train"] = dataset["train"].map(adjust_labels, batched=True, features=features)
```

```python
# Splitting the dataset into train, validation, and test sets
dataset = dataset["train"].train_test_split(test_size=0.2)
dataset_clean = dataset["train"].train_test_split(train_size=0.8, seed=42)
dataset_clean["validation"] = dataset_clean.pop("test")
dataset_clean["test"] = dataset["test"]
dataset = dataset_clean
```

```python
#Print the dataset to see the distribution of the sets
dataset
```

```python
# Import necessary libraries for fine-tuning a pre-trained model
from transformers import AutoModelForSequenceClassification, AutoTokenizer, DataCollatorWithPadding
from transformers import Trainer, TrainingArguments
```

```python
# Pre-trained model name or path
MODEL = f"mdraw/german-news-sentiment-bert"
```

```python
#Load the model
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
#Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(MODEL)
```

```python
# Tokenization function for the dataset
def tokenize_function(dataset):
    return tokenizer(dataset["text"],  padding=True)
# Tokenize the datasets
tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
#Check the tokenized datasets
tokenized_datasets
```

```python
#Import necessary libraries and modules for computing metrics
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import accuracy_score

import numpy as np
# Compute evaluation metrics
def compute_metrics(eval_preds):
    logits, labels = eval_preds
    predictions = np.argmax(logits, axis=-1)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, predictions, average='macro', zero_division=1)
    metric = {
        'accuracy': accuracy_score(labels, predictions),
        'precision': precision,
        'recall': recall,
        'f1': f1,
    }
    return metric
```

```python
# Create data collator for padding sequences
data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
```

```python
batch_size=6

# Set the training arguments for the Trainer
training_args = TrainingArguments(
    "test-sentiment",  # Output directory for the trained model
    learning_rate=1e-4,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    evaluation_strategy="epoch",  # Evaluate the model after each epoch
    save_strategy="epoch",  # Save the model after each epoch
    num_train_epochs=10,  # Number of training epochs
    weight_decay=0.00001,  # Weight decay for regularization
    logging_steps=1,  # Log metrics at every step
)
```

```python
# Create the Trainer object for training and evaluation
trainer = Trainer(
    model,  # The model to be trained
    training_args,  # Training arguments
    train_dataset=tokenized_datasets["train"],  # Training dataset
    eval_dataset=tokenized_datasets["validation"],  # Evaluation dataset
    data_collator=data_collator,  # Data collator for padding
    tokenizer=tokenizer,  # Tokenizer used for tokenization
    compute_metrics=compute_metrics  # Function for computing evaluation metrics
)
```

```python
#Train the model
trainer.train()
```

```python
# Save the trained model
model_path = "saved_final_model_sa"
model.save_pretrained(model_path)

from google.colab import files
import shutil

# Zip the saved model directory
shutil.make_archive('saved_final_model_sa', 'zip', 'saved_final_model_sa')

# Download the saved model zip file
files.download('saved_final_model_sa.zip')
```

```python
# Evaluate the model on the evaluation dataset
trainer.evaluate()
```

```python
#Evaluate on the test set
test_result = trainer.evaluate(tokenized_datasets["test"])

#Print the evaluation metrics
print(test_result)
```

```python
# Generate predictions on the test dataset and compute accuracy
predictions = trainer.predict(tokenized_datasets["test"])
preds = np.argmax(predictions.predictions, axis=-1)
metric = load_metric("accuracy")
metric.compute(predictions=preds, references=predictions.label_ids)
```
